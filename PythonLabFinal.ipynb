1. Hey!! We all know “Kung Fu Panda” and “FoodPanda” but
today's story is about “CodePanda”. Codepanda is a panda
who lives in a city named “PandaLand”. And she is the best
programmer in PandaLand. CodePanda’s little brother
“CoolPanda” is also a programmer. CodePanda has a
fantasy about collecting Bangladeshi Bank Notes. So she
always expects different types of bank notes as gifts and all
of her family members know that.
On her 17th birthday, her father has given her a full box of
Bangladeshi bank notes as presents.
Now, CoolPanda is requesting CodePanda to give him some notes from the box. CodePanda
has agreed to give him some notes but according to a condition. The condition is, CodePanda
will give him the notes from the list which index is divisible by 3.
Now both CodePanda and CoolPanda are thinking which notes they will get. As they are busy
unboxing, can you write a program to solve the question?
The sequence of notes kept inside the box is given below.
[10, 2, 10, 20, X, 2, X, 500, 1000, 2, 200, 200, 100, 5, 50, X, 1000, 2, 500, 20, 200, 500, 10,
200, 5, 100, 10, 50, X, 200, 500, 5, 1000, 100, 20, 1000, X, 500, 10, 20, 5, 20, 100, 50, X]
To make it hard for you, CodePanda has replaced the amount of some notes with X. You have
to replace X with a valid Note. You can consider the value of X= 2, 5, 10, 50, 100, 500, 1000.
Now, you have to write a Python program to find out which notes and how many of each note
both pandas will get.
Don’t worry. You just have to solve for any one of them. Students with an Even number as
the last digit of their ID(ex: 191_15_00000, 191_15_00002) will find out every specific note
and the amount of each note CodePanda will get.
Students with an Odd number as the last digit of their ID(ex: 191_15_00001) will find out
every specific note and the amount of each note CoolPanda will get.

Output should look like:
x1
- y1
x2
- y2
.
.
xn
- yn
Where x1
, x2
,...,xn are the values of the note and y1
, y2
,...,yn are the amount of notes.

```
all_notes = [10, 2, 10, 20, 5, 2, 5, 500, 1000, 2, 200, 200, 100, 5, 50, 5, 1000, 2, 500, 20, 200, 500, 10,200, 5, 100, 10, 50, 5, 200, 500, 5, 1000, 100, 20, 1000, 5, 500, 10, 20, 5, 20, 100, 50, 5]
coolpanda = []


for index,value in enumerate(all_notes):
    if index % 3 == 0:
        coolpanda.append(value)
    
print(coolpanda)

note2 = []
note5 = []
note10 = []
note20 = []
note50 = []
note100 = []
note200 = []
note500 = []
note1000 = []

for coolpanda_index,coolpanda_value in enumerate(coolpanda):
    if coolpanda_value == 2:
       note2.append(coolpanda_value)
    elif coolpanda_value == 5:
       note5.append(coolpanda_value)
    elif coolpanda_value == 10:
        note10.append(coolpanda_value)
    elif coolpanda_value == 20:
        note20.append(coolpanda_value)
    elif coolpanda_value == 50:
        note50.append(coolpanda_value)
    elif coolpanda_value == 100:
        note100.append(coolpanda_value)
    elif coolpanda_value == 200:
        note200.append(coolpanda_value)
    elif coolpanda_value == 500:
        note500.append(coolpanda_value)
    elif coolpanda_value == 1000:
        note1000.append(coolpanda_value)

print(note2)
print(note5)
print(note10)
print(note20)
print(note50)
print(note100)
print(note200)
print(note500)
print(note1000)


print(f"2 taka note : {sum(note2)}, 5 taka note : {sum(note5)}, 10 taka note : {sum(note10)}, 20 taka note : {sum(note20)}, 50 taka note : {sum(note50)},"
      f"100 taka note : {sum(note100)}, 200 taka note : {sum(note200)}, 500 taka note : {sum(note500)}, 1000 taka note : {sum(note1000)}")
```

2. CodePanda has grown up. And it’s her wedding!! Now she is sitting on a couch and watching
her relatives and family members are enjoying the ceremony. Some of her cousins are dancing,
some are singing, some are having adda time with their friends, some are eating and feeding
their childrens. Her uncles are also busy with different activities. Some of them are eating, some
gossiping, some of them are monitoring the whole hall. Her friends are with her. They all are
sitting with her.
CodePanda suddenly thought that it’s a interesting scenario. She can design OOP relations
from this scenario. Thinking about programming on her own wedding sounds a bit weird. Right?
Maybe it’s true for the rest of the world but not for the best programmer of PandaLand.
Programming is her one and only passion. Whatever, she cannot implement the scenario right
now, as she does not have her computer with her. But at the same time, she does not want to
miss the scenario. So, she asked you for a favor to implement the scenario. Can you do this for
your friend CodePanda?
You have to consider CodePanda’s each type of relatives/friends of the above scenario as a
class and each person as an object. Each class can have multiple objects and multiple objects
can do the same work. Please feel free to consider the number of objects.
Now implement the whole scenario in OOP concept using Python.

```
class cousins:
    def dancing(self):
        print("cousins are Dancing")

    def singing(self):
        print("cousins are Singing")

    def adda_time(self):
        print("cousins are Adda_time with their Friends")

    def eating(self):
        print("cousins are Eating Food")

    def feading (self):
        print("cousins are Feading their child")


class uncle:
    def gossiping(self):
        print("uncle are Gossiping")

    def monitoring(self):
        print("uncle are Monitoring the whole hall")

    def Eating(self):
        print("uncle are Eating Food")

class friend:
    def siting(self):
        print('Friends are siting with Codepanda')



class codepanda_wedding(cousins,uncle,friend):
    def wedding(self):
        print("IT's codepanda wedding")



obj=codepanda_wedding()
obj.wedding()
obj.dancing()
obj.eating()
obj.singing()
obj.adda_time()
obj.feading()
obj.gossiping()
obj.monitoring()
obj.Eating()
obj.siting()
```
